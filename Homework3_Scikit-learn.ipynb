```python
import numpy as np
import pandas as pd
```

```python
from sklearn.datasets import load_boston
```

```python
boston = load_boston()
boston.keys()
```

Разбейте эти датафреймы на тренировочные (X_train, y_train) и тестовые (X_test, y_test) с помощью функции train_test_split так, чтобы размер тестовой выборки составлял 30% от всех данных, при этом аргумент random state должен быть равен 42.

```python
data = boston["data"]
data.shape
```

```python
feature_names = boston["feature_names"]
feature_names
```

```python
target = boston["target"]
target[:10]
```

```python
X = pd.DataFrame(data, columns=feature_names)
X.head()
```

```python
X.info()
```

```python
y = pd.DataFrame(target, columns=["price"])
y.info()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

Создайте модель линейной регрессии под названием lr с помощью класса LinearRegression из модуля sklearn.linear_model.

Обучите модель на тренировочных данных (используйте все признаки) и сделайте предсказание на тестовых.

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
y_pred = lr.predict(X_test)
y_pred.shape
```

```python
check_test = pd.DataFrame({
    "y_test": y_test["price"],
    "y_pred": y_pred.flatten(),
})

check_test.head(10)
```

```python
#Оценим насколько реальные значения отличаются от предсказанных
check_test["error"] = check_test["y_pred"] - check_test["y_test"]
check_test.head()
```

```python
#Найдем метрику R2 для данной модели
from sklearn.metrics import r2_score
r2_score_1=r2_score(check_test["y_pred"], check_test["y_test"])
r2_score_1
```
